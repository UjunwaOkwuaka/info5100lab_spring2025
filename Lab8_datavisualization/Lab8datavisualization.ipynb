```python

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
 


df = pd.read_csv('Iris.csv')
 

df = df.drop(columns=['Id'], errors='ignore')  # drop 'Id' column if it exists
 

sns.set(style="whitegrid")
 
#Question 1- Histogram of Sepal Length by Species

plt.figure(figsize=(8, 5))

sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True, bins=20)

plt.title('Histogram of Sepal Length by Species')

plt.show()
 
#Question 2- Scatter Plot of Sepal Length vs. Sepal Width

plt.figure(figsize=(8, 5))

sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')

plt.title('Sepal Length vs. Sepal Width')

plt.show()
 
#Question 3- Box Plot of Petal Length for Each Species

plt.figure(figsize=(8, 5))

sns.boxplot(data=df, x='Species', y='PetalLengthCm')

plt.title('Box Plot of Petal Length by Species')

plt.show()
 
#Question 4- Pairplot of All Features Colored by Species

sns.pairplot(df.drop(columns='Id', errors='ignore'), hue='Species')

plt.suptitle('Pairplot of Iris Dataset Features', y=1.02)

plt.show()

 
```
